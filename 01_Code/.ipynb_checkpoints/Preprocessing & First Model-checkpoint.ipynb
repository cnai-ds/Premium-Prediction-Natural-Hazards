## Clean 'DEDUCTIBLES'

```python
df[['a', 'b']] = df.DEDUCTIBLES.str.split(', ',expand=True)
df['pd_bi'] = ""
df['pd'] = ""
df['bi'] = ""
```

```python
for i, value in enumerate(df.a):
    try:
        if value[:6] == 'PD,BI:':
            df['pd_bi'][i] = value[6:]
            df['pd'][i] = np.nan
        elif value[:3] == 'PD:' and value[9:12] != 'BI:':
            df['pd_bi'][i] = np.nan
            df['pd'][i] = value[3:]
    except Exception: # pass NaNs
        pass 
    
for i, value in enumerate(df.b):
    try:
        df['bi'][i] = value[4:]
    except Exception: # pass nans
        df['bi'][i] = np.nan
```

```python
df.drop(['a','b'], axis=1, inplace=True)
```

```python
# clean PD_BI
for i, value in enumerate(df['pd_bi']):
    try:
        df['pd_bi'][i] = df['pd_bi'][i].replace(',','.')
        df['pd_bi'][i] = df['pd_bi'][i].replace('M','')
        df['pd_bi'] = df['pd_bi'].apply(pd.to_numeric)
    except Exception:
        pass
```

```python
df['pd'].unique()
```

```python
for i, value in enumerate(df['pd']):
    try:
        if 'M' in value:
            df['pd'][i] = df['pd'][i].replace(' ' and 'M', '')
            df['pd'][i] = df['pd'][i].replace(',','.')
        elif 'loss' in value:
            df['pd'][i] = df['pd'][i].replace('% of loss' or '%ofloss', '')
            df['pd'][i] = df['pd'][i].replace(',','.')
            df['pd'][i] = pd.to_numeric(df['pd'][i])
            df['pd'][i] = df['pd'][i] / 100
        elif 'tiv' in value:
            df['pd'][i] = df['pd'][i].replace('% of tiv' or '%oftiv', '')
            df['pd'][i] = df['pd'][i].replace(',','.')
            df['pd'][i] = pd.to_numeric(df['pd'][i])
            df['pd'][i] = df['pd'][i] / 100  
    except Exception as e:
        pass
df['pd'] = df['pd'].apply(pd.to_numeric)
```

```python
df['pd'].unique()
```

```python
df['bi'].unique()
```

```python
# clean BI
for i, value in enumerate(df['bi']):
    try:
        df['bi'][i] = df['bi'][i].replace('M', '')
        df['bi'][i] = df['bi'][i].replace('Day(s)' and '% of loss', '')
        df['bi'][i] = df['bi'][i].replace('Day(s)' or '% of loss', '')
        df['bi'][i] = df['bi'][i].replace(',','.')
    except Exception as e:
        pass
```

```python
df['bi'] = df['bi'].apply(pd.to_numeric)
df['bi'].unique()
```

## Train model

```python
from fancyimpute import KNN
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
import xgboost
from xgboost import plot_importance
from sklearn.metrics import mean_squared_error
import joblib
import re
regex = re.compile(r"\[|\]|<", re.IGNORECASE)
```

```python
train, test = train_test_split(df, test_size = 0.15)
```

```python
knn_imputer = KNN()
train_impute = pd.DataFrame(KNN(k=5).fit_transform(train), columns=df.columns)
test_impute = pd.DataFrame(KNN(k=5).fit_transform(test), columns=df.columns)
```

```python
X_train = train_impute.drop(['fitted_PREMIUM'], axis=1)
X_train.columns = [regex.sub("_", col) if any(x in str(col) for x in set(('[', ']', '<'))) else col for col in X_train.columns.values]
y_train = train_impute['fitted_PREMIUM']
X_test = test_impute.drop(['fitted_PREMIUM'], axis=1)
X_test.columns = [regex.sub("_", col) if any(x in str(col) for x in set(('[', ']', '<'))) else col for col in X_test.columns.values]
y_test = test_impute['fitted_PREMIUM']
```

## Ramdom Forest

```python
rf = RandomForestRegressor(n_estimators = 800)
rf.fit(X_train, y_train)
y_rf = rf.predict(X_test)
```

```python
y_test_ = inv_boxcox(y_test, 0.17700863235182274)
y_rf_ = inv_boxcox(y_rf, 0.17700863235182274)
```

```python
rf_rmse = np.sqrt(mean_squared_error(y_test_, y_rf_))
rf_rmse
```

```python
rf_rmse = np.sqrt(mean_squared_error(y_test, y_rf))
print("RMSE: %f" % (rf_rmse))
```

```python
rf_nrmse = rf_rmse/(max(df['fitted_PREMIUM'])-min(df['fitted_PREMIUM']))
print("Normalized RMSE: %f" % (rf_nrmse))
```

## Xgboost

```python
xgb = xgboost.XGBRegressor(max_depth=5, n_estimators=800)
xgb.fit(X_train, y_train)
y_xgb = xgb.predict(X_test)
```

```python

y_test__ = inv_boxcox(y_test, 0.17700863235182274)
y_rf__ = inv_boxcox(y_xgb, 0.17700863235182274)
```

```python
xgb_rmse = np.sqrt(mean_squared_error(y_test__, y_rf__))
xgb_rmse
```

```python
xgb_rmse = np.sqrt(mean_squared_error(y_test, y_xgb))
print("RMSE: %f" % (xgb_rmse))
```

```python
xgb_nrmse = xgb_rmse/(max(df['fitted_PREMIUM'])-min(df['fitted_PREMIUM']))
print("Normalized RMSE: %f" % (xgb_nrmse))
```

```python
def feature_importance(model, X_columns):
    plt.figure(figsize=(10, 15))
    ordering = np.argsort(model.feature_importances_)[::1][150:]
    importances = model.feature_importances_[ordering]
    feature_names = X_columns[ordering]
    x = np.arange(len(feature_names))
    width = 0.4
    plt.barh(feature_names, importances, width)
    plt.title("Feature Importance" ,fontsize=12)
    plt.xlabel("Importances" ,fontsize=12)
    plt.ylabel("Features" ,fontsize=12)
    plt.rc('ytick', labelsize=12) 
```

```python
ordering = np.argsort(xgb.feature_importances_)[::1][:]
importances = xgb.feature_importances_[ordering]
feature_names = X_train.columns[ordering]
```

```python
importances = pd.DataFrame(importances, columns=['importances'])
feature_names = pd.DataFrame(feature_names, columns=['feature_names'])
```

```python
result = pd.concat([feature_names, importances], axis=1)
result[result['feature_names'].str.contains('MAINOCCUPANCY')].tail(30)
```

## Top feature importances

```python
fig, ax = plt.subplots(figsize=(15,15))
plot_importance(xgb, max_num_features=30, ax=ax)
```
