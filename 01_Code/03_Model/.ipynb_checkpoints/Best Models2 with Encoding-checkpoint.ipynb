```python
df = shuffle(df)
df.drop(['FACUL_NUM','DIVISION_NUM','UWYEAR','PREMIUM','INSUREDVALUEPD','INSUREDVALUEBI','TOTALINSUREDVALUE'], axis=1, inplace=True)
df = pd.get_dummies(df, drop_first=True, dummy_na=True)
```

```python
print(df.shape)
df.head()
```

## Impute missing data with KNN imputer

```python
train, test = train_test_split(df, test_size=0.15, random_state=42)
knn_imp = KNNImputer(n_neighbors=4, weights='distance')
train_impute = pd.DataFrame(knn_imp.fit_transform(train), columns=df.columns)
test_impute = pd.DataFrame(knn_imp.fit_transform(test), columns=df.columns)
```

```python
train_impute.head()
```

```python
X_train = train_impute.drop(['fitted_PREMIUM'], axis=1)
X_train.columns = [regex.sub("_", col) if any(x in str(col) for x in set(('[', ']', '<'))) else col for col in X_train.columns.values]
y_train = train_impute['fitted_PREMIUM']
X_test = test_impute.drop(['fitted_PREMIUM'], axis=1)
X_test.columns = [regex.sub("_", col) if any(x in str(col) for x in set(('[', ']', '<'))) else col for col in X_test.columns.values]
y_test = test_impute['fitted_PREMIUM']
```

# Model Training

```python
import re
regex = re.compile(r"\[|\]|<", re.IGNORECASE)
import xgboost
from xgboost import plot_importance
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from keras.layers import Dropout
from tensorflow.keras.layers.experimental import preprocessing
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.feature_selection import RFECV
import joblib
import eli5
from eli5.sklearn import PermutationImportance
```

## XGBoost

```python
xgb = xgboost.XGBRegressor(max_depth=4, n_estimators=550, eta=0.2, grow_policy='lossguide',
                           max_leaves=4, reg_lambda=0, reg_alpha=2, eval_metric=["rmse"])
xgb.fit(X_train, y_train)
pred_xgb = xgb.predict(X_test)
```

```python
y_test__ = inv_boxcox(y_test, 0.17700863235182274)
y_xgb__ = inv_boxcox(pred_xgb, 0.17700863235182274)
```

```python
y_test__.max()
```

```python
xgb_rmse = np.sqrt(mean_squared_error(y_test__, y_xgb__))
xgb_rmse
```

```python
xgb_rmse = np.sqrt(mean_squared_error(y_test, pred_xgb))
print("RMSE: %f" % (xgb_rmse))
xgb_nrmse = xgb_rmse/(max(df['fitted_PREMIUM'])-min(df['fitted_PREMIUM']))
print("Normalized RMSE: %f" % (xgb_nrmse))
```

```python
eval_set = [(X_train, y_train), (X_test, y_test)]
xgb.fit(X_train, y_train, eval_metric=["rmse"], eval_set=eval_set, verbose=False)
results = xgb.evals_result()
epochs = len(results['validation_0']['rmse'])
x_axis = range(0, epochs)

fig, ax = plt.subplots(figsize=(10,5))
ax.plot(x_axis, results['validation_0']['rmse'], label='Train')
ax.plot(x_axis, results['validation_1']['rmse'], label='Test')
ax.legend()
plt.xlabel('Number of estimators',fontsize=18)
plt.ylabel('RMSE',fontsize=18)
plt.title('XGBoost RMSE',fontsize=18)
plt.show()
```

```python
perm_xgb = PermutationImportance(xgb).fit(X_train, y_train)
weights_xgb = eli5.show_weights(perm_xgb, feature_names = X_train.columns.tolist())
weights_xgb
```

## Random Forest

```python
rf = RandomForestRegressor(max_depth=5, min_samples_leaf=3, n_estimators=1000)
rf.fit(X_train, y_train)
pred_rf = rf.predict(X_test)
```

```python
rf_rmse = np.sqrt(mean_squared_error(y_test, pred_rf))
print("RMSE: %f" % (rf_rmse))
rf_nrmse = rf_rmse/(max(df['fitted_PREMIUM'])-min(df['fitted_PREMIUM']))
print("Normalized RMSE: %f" % (rf_nrmse))
```

```python
perm_rf = PermutationImportance(rf).fit(X_train, y_train)
weights_rf = eli5.show_weights(perm_rf, feature_names = X_train.columns.tolist())
weights_rf
```

## DNN

```python
normalizer = preprocessing.Normalization(input_shape=[47,])
normalizer.adapt(np.array(X_train))
dnn = keras.Sequential([
    normalizer,
    keras.layers.Dense(40, input_dim=47, activation='relu'),
    keras.layers.Dense(35, activation='relu'),
    keras.layers.Dropout(0.3),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dropout(0.3),
    keras.layers.Dense(1)])
dnn.compile(loss='mean_absolute_error', optimizer='adam',metrics=[tf.keras.metrics.RootMeanSquaredError()])
```

```python
dnn.summary()
```

```python
fitted = dnn.fit(X_train, y_train, validation_split=0.10, verbose=0, epochs=200)
plot_loss(fitted)
```

```python
def create_model():
    model = keras.Sequential([
        normalizer,
        keras.layers.Dense(40, input_dim=43, activation='relu'),
        keras.layers.Dense(35, activation='relu'),
        keras.layers.Dropout(0.3),
        keras.layers.Dense(20, activation='relu'),
        keras.layers.Dropout(0.3),
        keras.layers.Dense(1)])
    model.compile(loss='mean_absolute_error', optimizer='adam',metrics=[tf.keras.metrics.RootMeanSquaredError()])
    return model
```

```python
pred_dnn = dnn.predict(X_test)
dnn_rmse = np.sqrt(mean_squared_error(y_test, pred_dnn))
print("RMSE: %f" % (dnn_rmse))
dnn_nrmse = dnn_rmse / (max(df['fitted_PREMIUM'])-min(df['fitted_PREMIUM']))
print("Normalized RMSE: %f" % (dnn_nrmse))
```

```python
dnn2 = KerasRegressor(build_fn=create_model)
dnn2.fit(X_train, y_train, epochs=200, verbose=0)
perm_dnn = PermutationImportance(dnn2).fit(X_train, y_train)
weights_dnn = eli5.show_weights(perm_dnn, feature_names = X_train.columns.tolist())
```
