```python
import sys
import os
import re
sys.path.append('../')
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
from datetime import datetime
import time
sns.set()
pd.set_option('display.max_columns', None)
pd.option_context("display.max_rows", None)
import warnings
warnings.filterwarnings("ignore")
from scipy import stats
import statsmodels.api as sm
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from scipy.special import inv_boxcox
from utils import *

from fancyimpute import KNN
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
import xgboost
from xgboost import plot_importance
from sklearn.metrics import mean_squared_error
import joblib
import re
regex = re.compile(r"\[|\]|<", re.IGNORECASE)
from sklearn.utils import shuffle
```

```python
df = pd.read_excel("../../01_Data/02_Processed/processed_v2.xlsx")
df.head()
```

```python
summary(df)
```

```python
for i, value in enumerate(df.MAIN_PRICING_CATEG):
    if value not in ['Mining - Hard Rock','Commercial buildings','Food producers']:
        df.MAIN_PRICING_CATEG[i] = 'Other'
    else:
        pass


for i, value in enumerate(df.MAIN_PRICING_SUBCATEG):
    if value not in ['Aggravated','Refinery - Standard',
                     'Low','Chocolate industry','Pure Pipeline',
                     'Combined cycle gas turbine power plant']:
        df.MAIN_PRICING_SUBCATEG[i] = 'Other'
    else:
        pass
    
for i, value in enumerate(df.SUBSECTOR):
    if value not in ['INTEGRATED CO,CONGLOMERATE','ONSHORE CO & CHEMICAL CO','MINING COMPANY',
                    'TRANSPORTATION & LOGISTICS','CORPORATE PROPERTY & CASUALTY','CONSUMER GOODS',
                     'ENERGY & NATURAL RESOURCES','POWER & UTILITY COMPANY']:
        df.SUBSECTOR[i] = 'Other'
    else:
        pass
```

```python
df.drop(['FACUL_NUM','DIVISION_NUM','UWYEAR','MARKET','MAINOCCUPANCY',
         'premium_level','ORIGCCY','WORDING','QUALITY_RISK_MGT','MB_QUALITY','MAIN_COUNTRY',
        'GUARANTEE','TAG','inception_year','COVERAGE'], axis=1,inplace=True)
df = pd.get_dummies(df, drop_first=True, dummy_na=True)
df.drop(['MAIN_PRICING_CATEG_Other','BI_MITIGATION_Average Plus', 'BI_MITIGATION_Good',
        'ASSET_QUALITY_Average Plus', 'ASSET_QUALITY_Good','BI_MITIGATION_nan','CT_STATUS_Finalized'], axis=1,inplace=True)
```

```python
df.shape
```

```python
df['fitted_PREMIUM'], lmbda = stats.boxcox(df['PREMIUM'])
sns.distplot(df['fitted_PREMIUM'], norm_hist=True, color = 'red')
df.drop(['PREMIUM'], axis=1, inplace =True)
df.shape
```

```python
df = shuffle(df)
```

```python
train, test = train_test_split(df, test_size = 0.12)
train_impute = pd.DataFrame(KNN(k=5).fit_transform(train), columns=df.columns)
test_impute = pd.DataFrame(KNN(k=5).fit_transform(test), columns=df.columns)
```

```python
X_train = train_impute.drop(['fitted_PREMIUM'], axis=1)
X_train.columns = [regex.sub("_", col) if any(x in str(col) for x in set(('[', ']', '<'))) else col for col in X_train.columns.values]
y_train = train_impute['fitted_PREMIUM']
X_test = test_impute.drop(['fitted_PREMIUM'], axis=1)
X_test.columns = [regex.sub("_", col) if any(x in str(col) for x in set(('[', ']', '<'))) else col for col in X_test.columns.values]
y_test = test_impute['fitted_PREMIUM']
```

```python
xgb = xgboost.XGBRegressor(max_depth=3, n_estimators=540, eta=0.2, grow_policy='lossguide',
                           max_leaves=4, reg_lambda=0, reg_alpha=3, eval_metric=["rmse"]).fit(X_train, y_train)
y_xgb = xgb.predict(X_test)
```

```python
y_test__ = inv_boxcox(y_test, 0.17700863235182274)
y_xgb__ = inv_boxcox(y_xgb, 0.17700863235182274)
```

```python
y_test__.max()
```

```python
xgb_rmse = np.sqrt(mean_squared_error(y_test__, y_xgb__))
xgb_rmse
```

```python
xgb_rmse = np.sqrt(mean_squared_error(y_test, y_xgb))
print("RMSE: %f" % (xgb_rmse))
```

```python
xgb_nrmse = xgb_rmse/(max(df['fitted_PREMIUM'])-min(df['fitted_PREMIUM']))
print("Normalized RMSE: %f" % (xgb_nrmse))
```
